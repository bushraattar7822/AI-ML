```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
gh=pd.read_csv("linear_regression_training_dataset_5000.csv")
print(gh)
print(gh.head())
print(gh.describe())
#define feautre(x) and target(y)
feature_cols=["ad_spend","store_visits","discount","competitor_price","season_index"]
x=gh[feature_cols]
y=gh["sales"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
                                               #for testing=20%  #for training=80%
#train the linear regression model
model=LinearRegression();model.fit(x_train,y_train)
print("\n---Model coefficient---")
for name, coef in zip(feature_cols,model.coef_):
    print(f"{name}:{coef:.4f}")
    print(f"Intercept:{model.intercept_:.4f}")

#Evaulate the model
y_pred=model.predict(x_test)
r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=mse**0.5
print("\n---Evaluation metrics on test set---")
print(f"R^2 score:{r2:.4f}")
print(f"MAE:{mae:.2f}")
print(f"MSE:{mse:.2f}")
print(f"RMAE:{rmse:.2f}")

#plot predicted value and actual value
plt.figure(figsize=(8,8))#figure size
plt.scatter(y_test,y_pred,alpha=0.5)#y_test:actual value y_pred:predicted value alpha:opaque/transparent
plt.xlabel("Actual sales")
plt.ylabel("Predicted sales")

#plot ideal line
min_val=min(y_test.min(),y_pred.min()) #minimum values of actual sales and predicted sales
max_val=max(y_test.max(),y_pred.max())#maximum value of actual sales and predictes sales
plt.plot([min_val,max_val],[min_val,max_val])
plt.tight_layout()
plt.show()
```
